```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
links = []
titles = []
prices = []
ratings = []
company = []
id = []
specs = []
reviews = []
rev_id = []
score = []

id_generator = 1
link_iterator = 0
for a in range (5):
    ad_grid = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[4]/div[1]/div/div[1]/div[2]')
    ad_cards = ad_grid.find_elements(By.CLASS_NAME, 'gridItem--Yd0sa')

    temp1 = ad_grid.find_elements (By.CLASS_NAME, 'title--wFj93')
    temp2 = ad_grid.find_elements (By.CLASS_NAME, 'price--NVB62')

    for i in range (40):
        if 'case' not in temp1 [i].text and 'cover' not in temp1 [i].text and 'charger' not in temp1 [i].text:
            titles.append (temp1 [i].text)
            prices.append (temp2 [i].text)
            link = ad_cards [i].find_element (By.TAG_NAME, 'a')
            links.append (link.get_attribute ('href'))
        else:
            continue

    for i in range (len (ad_cards)):
        driver.get (links [link_iterator])
        id.append (id_generator)
        id_generator += 1
        link_iterator += 1 

        try:
            company.append (driver.find_element (By.XPATH, '/html/body/div[4]/div/div[3]/div[2]/div/div[1]/div[5]/div/a[1]').text)
        except NoSuchElementException:
            company.append ('-')
            pass

        driver.execute_script("window.scrollTo(0, 500);")
        time.sleep(1)

        try:
            ratings.append (driver.find_element (By.XPATH, '/html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[2]/div[1]/div[1]/div[1]/span[1]').text)
        except NoSuchElementException:
            ratings.append ('-')
            pass

        driver.execute_script("window.scrollTo(500, 1200);")

        temp3 = []
        temp4 = []
        temp5 = []
        for j in range (5):
            try:
                grid = driver.find_element(By.XPATH, '/html/body/div[4]/div/div[9]/div[1]/div[1]/div/div/div/div[3]')
            except NoSuchElementException:
                break

            all = grid.find_elements (By.CLASS_NAME, 'review-item')

            for k in range (len (all)):
                try:
                    temp3.append (all [k].find_element (By.CLASS_NAME, 'user').text)
                except NoSuchElementException:
                    continue
                try:
                    temp4.append (all [k].find_element (By.CLASS_NAME, 'review-content-sl').text)
                except NoSuchElementException:
                    continue
                extra = all [k].find_element (By.CLASS_NAME, 'container-star ')
                star = extra.find_elements (By.TAG_NAME, 'img')
                r = 0
                for l in range (len (star)):
                    if star [l].get_attribute ('src') == "https://laz-img-cdn.alicdn.com/tfs/TB19ZvEgfDH8KJjy1XcXXcpdXXa-64-64.png":
                        r += 1
                temp5.append (r)
            
            if(j != 4):
                try:
                    nextButton = driver.find_element(By.CLASS_NAME,'ant-pagination-next')
                    next = nextButton.find_element(By.TAG_NAME,'button')
                    try:
                        next.click()
                        time.sleep (1)
                    except ElementClickInterceptedException:
                        break
                except NoSuchElementException:
                    break

        rev_id.append (temp3)
        reviews.append (temp4)
        score.append (temp5)
        
        driver.execute_script("window.scrollTo(1200, 1500);")
        
        try:
            specs.append (driver.find_element (By.XPATH, '/html/body/div[4]/div/div[9]/div[1]/div[2]/div/div/div[1]/div[1]/ul/li').text)
        except NoSuchElementException:
            specs.append ('-')
            pass
        
        driver.back ()
        time.sleep (1)
        
    if a != 4:
        next_button = driver.find_element (By.CSS_SELECTOR, 'li[title="Next Page"]')
        next_button.click ()
        time.sleep (3)
```
